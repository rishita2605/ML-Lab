```python
from collections import Counter
import pandas as pd

df_tennis = pd.read_csv('PlayTennis.csv')
df_tennis
```

```python
attr = list(df_tennis.columns)
print(attr)
attr.remove('PlayTennis')
print('Predicting attr', attr)
```

```python
import math

def entropy(probs):
  return sum ([-prob*math.log(prob, 2) for prob in probs])

def entropy_of_list(yes_no_list):
  class_count = Counter(x for x in yes_no_list)
  instance_len = len(yes_no_list)
  probablities = [x/instance_len for x in class_count.values()]
  return entropy(probablities)
```

```python
def information_gain(df, split_attr, target_attr):
  df_split = df.groupby(split_attr)
  datalen = len(df.index)

  df_agg = df_split.agg({target_attr:[entropy_of_list, lambda x: len(x)/datalen]})[target_attr]
  df_agg.columns = ['Entropy', 'Proportion']

  new_entropy = sum(df_agg['Entropy'] * df_agg['Proportion'])
  old_entropy = entropy_of_list(df[target_attr])

  return old_entropy - new_entropy
```

```python
def id3(df, target_attr, attr, default_class=None, default_attr='S'):
  pos_neg = Counter(x for x in df[target_attr])
  if(len(pos_neg)==1):
    return next(iter(pos_neg))
  elif df.empty or (not attr):
    return default_class
  else:
    default_class = max(pos_neg.keys()) #does not really make a difference
    gains={}

    for a in attr:
      gains[a] = information_gain(df, a, target_attr)
    
    best_attr = max(gains, key= lambda x: gains[x])
    tree = {best_attr:{}}
    attr.remove(best_attr)

    for attr_val, data in df.groupby(best_attr):
      subtree = id3 (data, target_attr, attr, default_class)
      tree[best_attr][attr_val] = subtree
    
    return tree
```

```python
tree = id3(df_tennis, 'PlayTennis', attr)
print(tree)
```
